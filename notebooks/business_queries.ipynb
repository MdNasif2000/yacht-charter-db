# List the total length of stay, between two given dates of yachts in the fleet in each port.

```python
%%sql
SELECT YACHT.home_port, PORT.p_name, SUM(VISIT.visit_duration) AS Total_Stay
FROM YACHT 
JOIN CHARTER ON YACHT.y_name = CHARTER.y_name
JOIN VISIT  ON CHARTER.charter_id = VISIT.charter_id
JOIN PORT ON VISIT.p_name = PORT.p_name
WHERE VISIT.start_date BETWEEN '2018-01-22' AND '2018-10-29'
GROUP BY YACHT.home_port, PORT.p_name;
```

# Get a list of yachts (by name) visiting their home port between two given dates, together with the date of arrival and the length of the stay.

```python
%%sql
SELECT YACHT.y_name, VISIT.date_of_arrival, SUM(VISIT.visit_duration) AS Length_of_Stay
FROM YACHT 
JOIN CHARTER  ON YACHT.y_name = CHARTER.y_name
JOIN VISIT  ON CHARTER.charter_id = VISIT.charter_id
WHERE YACHT.home_port = VISIT.p_name
AND VISIT.date_of_arrival BETWEEN '2018-01-22' AND '2018-09-29'
GROUP BY YACHT.y_name, VISIT.date_of_arrival;
```

# List the ports visited by a given customer together with the date of arrival and length of stay, ordered by date.

```python
%%sql
SELECT VISIT.p_name, VISIT.date_of_arrival, VISIT.visit_duration AS Length_of_Stay
FROM CUSTOMER 
JOIN CHARTER  ON CUSTOMER.cx_id = CHARTER.cx_id
JOIN VISIT ON CHARTER.charter_id = VISIT.charter_id
WHERE CUSTOMER.cx_id = 'D17-022'
ORDER BY VISIT.date_of_arrival;
```

# Remove a yacht temporarily for a period of time (e.g., for servicing) and get a list of yachts that are not available
# Create a temporary table with 'available' status:

```python
%%sql
CREATE TEMPORARY TABLE temp_yacht as
SELECT *, 'available' as status 
FROM YACHT;
```

```python
%%sql
ALTER TABLE temp_yacht MODIFY COLUMN status VARCHAR(20);
```

# Update yacht status to 'unavailable' for a specific yacht:

```python
%%sql
UPDATE temp_yacht
SET status = 'unavailable'
WHERE y_name = 'Escape';
```

# Display the contents of the temporary table:

```python
%%sql
select *
from temp_yacht;
```

# Display the unavailable yachts from the temporary table:

```python
%%sql
SELECT *
FROM temp_yacht
WHERE status = 'unavailable';
```
